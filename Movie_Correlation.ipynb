# Budget vs Gross Correlation

```python
# First let's import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# read-in the data
df = pd.read_csv(r'C:\Users\balta\OneDrive\Desktop\Phyton Projects\DataCleaning\movies.csv')
```

```python
df.head()
```

```python
#Checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# removing missing data
df = df.dropna()
```

```python
df.dtypes
```

```python
# Changing data types
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# get the year from realeased column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Scatterplot budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# budget vs gross plot using seaborn
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.show()
```

```python
# look at correlation 
df.corr(numeric_only=True)
```

```python
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True, cbar=True)
plt.title('Correlation Matrix')
plt.show()
```
